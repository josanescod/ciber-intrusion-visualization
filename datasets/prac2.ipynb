```python
import pandas as pd
import numpy as np

# load dataset
dataset = 'cybersecurity_intrusion_data.csv'
df = pd.read_csv(dataset)

# show rows and columns
nrows, ncolumns = df.shape

print (f"El dataset té {ncolumns} columns.")
print (f"EL dataset té {nrows} files.")
```

```python
# Create new variables
import numpy as np
import pandas as pd

# packet_rate = mida del paquet / duració de la sessió
df['packet_rate'] = df['network_packet_size'] / df['session_duration']

# failed_login_ratio = logins fallats / intents de login
df['failed_login_ratio'] = np.where(
    df['login_attempts'] > 0,
    df['failed_logins'] / df['login_attempts'],
    0.0
)

# risk_level (categórica) combinant diversos indicadors
def compute_risk(row):
    if row['attack_detected'] == 1:
        return 'Confirmed'                     # atac ja etiquetat 
    if row['ip_reputation_score'] < 0.3 and row['unusual_time_access'] == 1:
        return 'High'                          # IP dolenta + horari rar
    if row['ip_reputation_score'] >= 0.3 and row['unusual_time_access'] == 0:
        return 'Low'                           # IP bona + horari normal
    return 'Medium'                           # cas intermedi

df['risk_level'] = df.apply(compute_risk, axis=1)

# encryption_strength – número que indica la fortalesa del xifratge
enc_map = {'AES': 2, 'DES': 1, None: 0, 'None': 0}
df['encryption_strength'] = df['encryption_used'].map(enc_map).fillna(0).astype(int)

# *** ADD THIS LINE TO ACTUALLY SAVE THE FILE ***
output_path = 'cybersecurity_intrusion_data_enriched.csv'
df.to_csv(output_path, index=False)
print(f"cybersecurity_intrusion_data_enriched.csv file created.")
```

https://en.wikipedia.org/wiki/Winsorizing --> podem aplicar aquesta tècnica estadística per reduir l'impacte dels valors atípics en un conjunt de dades

Cadria normalitzar dades? en aquest dataset concret no seria necessari

```python
# aplicació de winsorizing
from scipy.stats.mstats import winsorize

df_w = df2.copy()

cols = ['session_duration', 'packet_rate', 'failed_login_ratio']

for col in cols:
    df_w[col] = winsorize(df2[col], limits=[0.01, 0.01])
```
